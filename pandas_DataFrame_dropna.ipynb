```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
df
```

```python
df.dropna()
```

```python
df.dropna(axis='columns')
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(subset=['name', 'toy'])
```

```python
df.dropna(inplace=True)
df
```
